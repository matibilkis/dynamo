```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import curdoc
from bokeh.layouts import row

output_notebook()
curdoc().theme = 'dark_minimal'

S = np.load("results/100/0.4/Psuc.npy")
```

```python
from bokeh.palettes import Dark2_5 as palette
import itertools  
colors = itertools.cycle(palette) 
```

```python
# create a Figure object
p = figure(width=1000, height=1000)
etas_min = np.linspace(1e-10,.5,100)
whole_etas = np.concatenate([etas_min, (1-etas_min)[::-1]])


for k, c in zip(range(100), colors):
    p.line(whole_etas,S[k], alpha=0.75,color=c)

p.xgrid.visible = False
p.ygrid.visible = False

show(row(p), notebook_handle=True)
```
