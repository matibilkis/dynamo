```python
import numpy as np#
import matplotlib.pyplot as plt
```

```python
amps = np.arange(0.01,1,.1)
tot_L = 4
ells = range(1,tot_L)
ps = np.zeros((tot_L, len(amps), 200))
for indL in ells:
    for inda, a in enumerate(amps):
        ps[indL, inda] = np.load("/data/uab-giq/scratch/matias/dynamo/{}/{}/Psuc.npy".format(indL,np.round(a,3)))[0]
```

```python
def hel(alpha):
    return (1+np.sqrt(1-np.exp(-4*alpha**2)))/2
```

```python
colors = {0:"red",1:"blue",2:"pink",3:"black"}
for layer in range(1,tot_L):
    aa = [ps[layer,i,99] for i in range(len(amps))]
    plt.plot(amps,aa, label="L="+str(layer), color=colors[layer])
plt.plot(amps,1-hel(amps),'--')
plt.legend()
```
